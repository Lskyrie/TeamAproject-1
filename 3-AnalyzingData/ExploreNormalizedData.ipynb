```python
def year_recode (series):
    if series == 1999:
        return 1
    if series == 2000:
        return 2
    if series == 2001:
        return 3
    if series == 2002:
        return 4
    if series == 2003:
        return 5
    if series == 2004:
        return 6
    if series == 2005:
        return 7
    if series == 2006:
        return 8
    if series == 2007:
        return 9
    if series == 2008:
        return 10
    if series == 2009:
        return 11
    if series == 2010:
        return 12
    if series == 2011:
        return 13
    if series == 2012:
        return 14
    if series == 2013:
        return 15
    if series == 2014:
        return 16
    if series == 2015:
        return 17
    if series == 2016:
        return 18
    if series == 2017:
        return 19
    if series == 2018:
        return 20
    if series == 2019:        
        return 21
```

```python
Explore_data['YearR'] = Explore_data['Year'].apply(year_recode)
```
