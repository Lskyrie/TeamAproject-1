# 1 Cleaning the Datasets

```python
CropsProduction.drop(['Domain Code'], axis=1, inplace=True)
CropsProduction.drop(['Domain'], axis=1, inplace=True)
CropsProduction.drop(['Area Code (FAO)'],axis=1, inplace=True)
CropsProduction.drop(['Area'], axis=1, inplace=True)
CropsProduction.drop(['Element Code'], axis=1, inplace=True)
CropsProduction.drop(['Item Code (FAO)'], axis=1, inplace=True)
CropsProduction.drop(['Year Code'],axis=1, inplace=True)
CropsProduction.drop(['Flag'], axis=1, inplace=True)
CropsProduction.drop(['Flag Description'], axis=1, inplace=True)
CropsProduction.drop(['Element'], axis=1, inplace=True)
CropsProduction.head()
```

```python
ValueOfCrops.drop(['Domain Code'], axis=1, inplace=True)
ValueOfCrops.drop(['Domain'], axis=1, inplace=True)
ValueOfCrops.drop(['Area Code (FAO)'],axis=1, inplace=True)
ValueOfCrops.drop(['Area'], axis=1, inplace=True)
ValueOfCrops.drop(['Element Code'], axis=1, inplace=True)
ValueOfCrops.drop(['Item Code (FAO)'], axis=1, inplace=True)
ValueOfCrops.drop(['Year Code'],axis=1, inplace=True)
ValueOfCrops.drop(['Flag'], axis=1, inplace=True)
ValueOfCrops.drop(['Flag Description'], axis=1, inplace=True)
ValueOfCrops.drop(['Element'], axis=1, inplace=True)
ValueOfCrops.head()
```

```python
TempChanges.head()
```

```python
TempChanges.drop(['Domain Code'], axis=1, inplace=True)
TempChanges.drop(['Domain'], axis=1, inplace=True)
TempChanges.drop(['Area Code (FAO)'],axis=1, inplace=True)
TempChanges.drop(['Area'], axis=1, inplace=True)
TempChanges.drop(['Element Code'], axis=1, inplace=True)
TempChanges.drop(['Months Code'],axis=1, inplace=True)
TempChanges.drop(['Months'],axis=1, inplace=True)
TempChanges.drop(['Year Code'],axis=1, inplace=True)
TempChanges.drop(['Flag'], axis=1, inplace=True)
TempChanges.drop(['Flag Description'], axis=1, inplace=True)
TempChanges.drop(['Element'], axis=1, inplace=True)
TempChanges.head()
```

```python
ValueOfCrops.rename(columns={'Value': 'US$1000', 'Item': 'Crops'}, inplace=True)
ValueOfCrops.drop(['Unit'], axis=1, inplace=True)
ValueOfCrops.head()
```

```python
CropsProduction.rename(columns={'Value': 'Tonnes', 'Item': 'Crops'}, inplace=True)
CropsProduction.drop(['Unit'], axis=1, inplace=True)
CropsProduction.head()
```

```python
TempChanges.rename(columns={'Value': 'TempChange', 'Item': 'Crops'}, inplace=True)
TempChanges.drop(['Unit'], axis=1, inplace=True)
TempChanges.head()
```

# Merging Crops Production and Value of Crops Datasets


```python
CropsTotalOutput = pd.merge(CropsProduction, ValueOfCrops, how='inner', left_on=['Crops','Year'], right_on=['Crops','Year'])
CropsTotalOutput.head()
```

# Merging Crops Total Output and Temperature Changes Datasets

```python
CropsMergedData= pd.merge(CropsTotalOutput, TempChanges, how='inner', on='Year')
CropsMergedData.head()
```
