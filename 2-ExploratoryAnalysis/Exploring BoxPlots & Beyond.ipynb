```r
library(ggplot2)
install.packages("dplyr")
library(dplyr)
```

```r
explore_dataset <-read.csv('/Users/home/Desktop/Final Project/crops_recode_dataset.csv')
```

```r
head(explore_dataset)
```

```r
## Creating Scatterplots
```

```r
## Scatterplot temp change over years
```

```r
d <- ggplot(explore_dataset, aes(x = YearR, y = TempChange))
d + geom_point()
```

```r
## Creating Box Plots
```

```r
## boxplot Tonnes
```

```r
d <- ggplot(explore_dataset, aes(x = "", y = Tonnes))
d + geom_boxplot() + xlab("")
```

```r
ggplot(explore_dataset, aes(x = factor(YearR), y = Tonnes)) + geom_boxplot()
```

```r
ggplot(explore_dataset, aes(x = factor(YearR), y = Tonnes)) + geom_boxplot() + scale_y_log10()
```

```r
## boxplot USD1000
```

```r
d <- ggplot(explore_dataset, aes(x = "", y = USD1000))
d + geom_boxplot() + xlab("")
```

```r
ggplot(explore_dataset, aes(x = factor(YearR), y = USD1000)) + geom_boxplot()
```

```r
## boxplot TempChange
```

```r
d <- ggplot(explore_dataset, aes(x = "", y = TempChange))
d + geom_boxplot() + xlab("")
```
